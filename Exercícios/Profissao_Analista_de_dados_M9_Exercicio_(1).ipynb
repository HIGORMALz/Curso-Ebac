````
from openpyxl import load_workbook
planilhas = load_workbook(filename='credito.xlsx')
planilha = planilhas.active
````

* **Extraindo coluna Id**

````
id_arquivo = list()

cabecalho = next(planilha.values)
indice_coluna = cabecalho.index('id')
id_arquivo = [linha[indice_coluna] for linha in planilha.values if linha[indice_coluna] != 'id']

print(id_arquivo)
````

* **Extraindo coluna Sexo**

````
sexo_cliente = list()

cabecalho = next(planilha.values)
indice_coluna = cabecalho.index('sexo')
sexo_cliente = [linha[indice_coluna] for linha in planilha.values if linha[indice_coluna] != 'sexo']

print(sexo_cliente)
````

* **Extraindo coluna Idade**

````
idades = list()

cabecalho = next(planilha.values)
indice_coluna = cabecalho.index('idade')
idades = [linha[indice_coluna] for linha in planilha.values if linha[indice_coluna] != 'idade']

print(idades)
````

* **Extraindo coluna dos clientes inadimplentes**

````
clientes_inadimpentes = list()

cabecalho = next(planilha.values)
indice_default = cabecalho.index('default')
clientes_inadimpentes = [linha[indice_default] for linha in planilha.values if linha[indice_default] != 'default' and linha[indice_default] != 0]

print(clientes_inadimpentes)
````

* **Extraindo coluna dos clientes com estado civil == 'solteiro'**

````
estado_civil_cliente = list()

cabecalho = next(planilha.values)
indice_coluna = cabecalho.index('estado_civil')
estado_civil_cliente = [linha[indice_coluna] for linha in planilha.values if linha[indice_coluna] != 'estado_civil' and linha[indice_coluna] != 'na']

print(estado_civil_cliente)
````

## 2\. Excel para JSON

Como preparação para o próximo módulo, vamos trabalhar com o JSON, um formato semi-estruturado, muito utilizado em transmissão de dados da web e equivalente a um **dicionário** Python.

Utilizando o pacote Python `openpyxl` visto em aula, extraia os dados das colunas `escolaridade` e `tipo_cartao`, removendo duplicados. Com os dados, construa o dicionário Python `credito` com a seguinte estrutura:

```python
credito = {
  'tipo_cartao': ['silver', 'blue', 'gold', 'platinum'],
  'escolaridade': ['doutorado', 'mestrado', 'na', 'sem educacao formal', 'graduacao', 'ensino medio']
}
```

Para finalizar, utilize o código abaixo para converter o dicionário `credito` no formato JSON:

```python
import json

credito_json = json.dumps(credito, indent=4)
print(credito_json)
```

Extraia os números de contato do arquivo texto `ebac.txt` e salve-os no arquivo csv `contato_ebac.csv` com o separador `;` no seguinte formato:

```
tipo;numero
whatsapp;+551142002991
telefone;+551130303200
```

**Nota:** Escreva o código da sua solução abaixo em uma ou mais células, você não precisa enviar o arquivo csv gerado.

````
with open(file='./ebac.txt', mode='r', encoding='utf8') as arquivo:
  linhas = arquivo.readlines()

print(linhas)
````

````
linhas = filter(lambda linha: linha != '\n', linhas)
linhas = map(lambda linha: linha.strip(), linhas)
linhas = list(linhas)

print(linhas)
````

````
linhas_com_numeros = filter(lambda linha: '-' in linha, linhas)
linhas_com_numeros = list(linhas_com_numeros)

print(linhas_com_numeros)
````
