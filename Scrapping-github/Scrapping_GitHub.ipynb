```python
import requests
from bs4 import BeautifulSoup

requisicao = requests.get('https://github.com/trending')

# salvando o conteudo na variavel conteudo
conteudo = requisicao.content

github = BeautifulSoup(conteudo, 'html.parser')
```

```python
todos_projetos = github.find_all('article', attrs={'class': 'Box-row'}) # extraindo o conteudo html contendo todos os projetos
```

```python
lista_github = list()
for c in todos_projetos:
  # nome dos projetos
  nome_projetos = c.find('h2', attrs={'class': 'h3 lh-condensed'})
  nome_projetos = nome_projetos.text
  nome_projeto_git = nome_projetos.split("\n")
  nome_projeto_limpo = [' '.join(i.split()) for i in nome_projeto_git]
  saida = '\n'.join(nome_projeto_limpo)
  saida_limpa = saida.replace("\n", " ").strip()


  # estrelas totais
  estrelas = c.find('a', attrs={'class': 'Link Link--muted d-inline-block mr-3'})
  estrelas = estrelas.text # conteudo sem as tags html
  # removendo as quebras de linha e removendo os espaços
  total_estrelas = estrelas.replace("\n", " ").strip()


  # estrelas do dia
  estrelas_dia = c.find('span', attrs={'class': 'd-inline-block float-sm-right'})
  estrelas_dia = estrelas_dia.text # conteudo sem as tags html
  # removendo as quebras de linha e removendo os espaços
  apenas_estrelas = estrelas_dia.replace("\n", " ").strip().split()


  # compartilhamentos 'forks'
  for forks in c.find_all('a', attrs={'class': 'Link Link--muted d-inline-block mr-3'}):
    compartilhamento = forks.text
    compartilhamento_limpo = compartilhamento.replace("\n", " ").strip()

  # linguagem
  linguagem = c.find('span', attrs={'class': 'd-inline-block ml-0 mr-3'})
  if linguagem: # se existir a linguagem extraia a linguagem
    linguagem = linguagem.text
    linguagem_limpa = linguagem.replace("\n", " ").strip()

    # adicionando os dados extraidos na lista 'lista_github'
    lista_github.append([saida_limpa, total_estrelas, apenas_estrelas[0], compartilhamento_limpo, linguagem_limpa])
  else:
    lista_github.append([saida_limpa, total_estrelas, apenas_estrelas[0], compartilhamento_limpo, ' '])

print(lista_github)
```
